# PLOTLY ASSIGNMENT:

```python
#1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a three dimensional space.
 
np.random.seed(30)
data = {
'X': np.random.uniform(-10, 10, 300),
'Y': np.random.uniform(-10, 10, 300),
'Z': np.random.uniform(-10, 10, 300)
}

df = pd.DataFrame(data)

fig = go.Figure()
fig.add_trace(go.Scatter3d(x = df['X'],
                           y = df['Y'],
                           z= df['Z'], 
                           mode= 'markers'))
```

```python
#2. Using the Student Grades, create a violin plot to display the distribution of scores across different grade categories.
np.random.seed(15)
data = {
'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
'Score': np.random.randint(50, 100, 200)
}

df = pd.DataFrame(data)

fig = go.Figure()
fig.add_trace(go.Violin(x = df['Grade'], 
                        y= df['Score']))
fig.update_layout(title='distribution of scores')
fig.show()
```

```python
#Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.

np.random.seed(20)
data = {
'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
'Day': np.random.choice(range(1, 31), 100),
'Sales': np.random.randint(1000, 5000, 100)
}

df = pd.DataFrame(data)

fig = go.Figure()
fig.add_trace(go.Heatmap(z=df['Sales'], x=df['Day'], y=df['Month']))
fig.update_layout(title='Sales Heatmap by Month and Day')
fig.show()
```

```python
#4. Using the given x and y data, generate a 3D surface plot to visualize the function

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
'X': x.flatten(),
'Y': y.flatten(),
'Z': z.flatten()
}

Tdf = pd.DataFrame(data)

fig = go.Figure()

fig.add_trace(go.Scatter3d(x= Tdf['X'], 
                           y= Tdf['Y'],
                           z = Tdf['Z'],
                           mode='markers',
                         ))
```

```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.


# x = [1,2,3,4]
# y = [3,8,4,7]

x = np.linspace(0, 10, 1000)
y = np.sin(x)

output_file('text.html')
p = figure(title = 'sine wave')
p.xaxis.axis_label='x'
p.yaxis.axis_label='sin(x)'
# p.circle(x,y)
p.line(x,y)
show(p)
```

```python
#2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

a = 100
x = np.random.rand(a)
y = np.random.rand(a)
random_sizes = np.random.randint(10, 100, a)
random_color = np.random.choice(['red', 'blue', 'green'], a)

new_data_frame = pd.DataFrame({'x': x, 'y': y, 'size': random_sizes, 'Color': random_color})
new_data_frame

p = figure(title = 'Scatter plot randomly generated x and y values')
p.scatter(x,y, size = random_sizes, fill_color = random_color, legend_label = 'Here is the scatter_plot with random_data')

show(p)
```

```python
#3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

x=['Apples', 'Oranges', 'Bananas', 'Pears']

p = figure(x_range=x, title="Fruit Counts")
p.xaxis.axis_label= "Fruits"
p.yaxis.axis_label="Count"
p.vbar(x=x, top=[20, 25, 30, 35], width=0.5, bottom=0, color="green", legend_label="Fruit Counts") 
show(p)
```

```python
#4. Create a Bokeh histogram to visualize the distribution of the given data.

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# hist, edges = np.histogram(x, density=True, bins=bins)
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
         fill_color="skyblue", line_color="white",
         legend_label="Bokeh histogram")

show(p)
```
